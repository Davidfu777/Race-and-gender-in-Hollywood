```python
df_race = df[['4. Sex (Sheet11)', '6. Age (Sheet11)','5. Race (Sheet11)']]
df_race = df_race.rename(columns= {'4. Sex (Sheet11)': 'sex', '6. Age (Sheet11)': 'age', '5. Race (Sheet11)': 'race' })
```

```python
df_race.head()
```

```python
df_race['race'].unique()
```

```python
df_mVF_race = pd.crosstab(index=df_race['sex'], columns=df_race['race'])
df_mVF_race.head()
```

```python
df_male_race = df_race[df_race['sex']== 'Male']
df_male_race_final = pd.crosstab(index=df_male_race['age'], columns=df_male_race['race'])
df_male_race_final.head()
```

```python
df_female_race = df_race[df_race['sex']== 'Female']
df_female_race_final = pd.crosstab(index=df_female_race['age'], columns=df_female_race['race'])
df_female_race_final.head()
```

```python
df_male_race_final.to_csv('Male-Race-vs-Age.csv')
df_female_race_final.to_csv('Female-Race-vs-Age.csv')
df_mVF_race.to_csv('Gender-Race-and-Age.csv')
```
